```python
# fill NA values 
df["Age"]= df["Age"].fillna(df["Age"].mean())
```

```python
# get the survival rate based on age 
df.Age = df.Age.astype(int)
ages = df[df['Survived'] == 1]['Age'].sort_values()
dc =  {}
for age in ages:
    if age not in dc.keys():
        dc[age] = 1
    else:
        dc[age] +=1
print(dc)

plt.figure(figsize=(30,10))
key = list(dc.keys())
value = list(dc.values())
for index in range(len(key)):
    plt.bar(key[index],value[index],color ='maroon')
    plt.text(key[index],value[index],str(value[index]),color="green")
plt.xticks(np.arange(len(key)),key)
plt.title("Different Ages of number of people survived")
plt.xlabel("Age")
plt.ylabel("Number of people")
plt.show()
```

```python
# sort the data 

dc_sorted = sorted(dc.items(), key=lambda x: x[1], reverse=True)

print(dc_sorted)
```

```python
# age distribution of survived poeple

df[df['Survived'] == 1]['Age'].hist()
plt.title('Age discribution of survived people')
plt.xlabel('Age')
plt.ylabel('# of people')
plt.show()
```

```python
# replace male and female 
df['Sex'] = df['Sex'].replace({"female": 0, "male": 1})
df.head()
```

```python
males = df[(df["Survived"]==1) & (df.Sex==1)]["Sex"].count()
female = df[(df["Survived"]==1) & (df.Sex==0)]["Sex"].count()
value = [males,female]
labels = ["males","Females"]
plt.bar(np.arange(len(value)),value);
for index in range(len(value)):
    plt.text(index,value[index],str(value[index]),color="green")
plt.xticks(np.arange(len(labels)),labels)
plt.title("Comparison between number of males and females survived")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.style.use("ggplot")
plt.show()
```

```python
# number of survived people based on the class 
class_1 = df[df.Pclass==1].count()[0]
class_2 = df[df.Pclass==2].count()[0]
class_3 = df[df.Pclass==3].count()[0]
classs = [class_1,class_2,class_3]
plt.bar(df.Pclass.unique(),classs)
plt.xticks(np.arange(5),["","class 1","class 2","class 3",""])
plt.title("number of Survived people based on the class")
plt.xlabel("Class")
plt.ylabel("Numper of people")
plt.show()
```

```python
# drop the na values 
df.dropna(inplace=True)
print(df.Embarked.unique())
df.Embarked = df.Embarked.replace({"S":0, "C":1,"Q":2})
print(df.Embarked.isnull().sum())
print(df.shape)
```

# Modeling

```python
print(df)
```

```python
 from sklearn.model_selection import train_test_split
```

```python
X = df.drop("Survived", axis=1)
y = df["Survived"]

# test train split 
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)

print(X_train.shape)
```

## Logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_model = LogisticRegression()

logistic_model.fit(X_train,y_train)
```

```python
logistic_model.score(X_test,y_test)
```

```python
y_predict = logistic_model.predict(X_test)
y_predict
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_predict))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_predict)
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier();

model.fit(X_train,y_train)

y_predict = model.predict(X_test)

model.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier();

model.fit(X_train,y_train)

y_predict = model.predict(X_test)

model.score(X_test,y_test)
```
